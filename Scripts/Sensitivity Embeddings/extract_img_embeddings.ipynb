# Extract image embeddings (VGG16)

Embed the stimulus images for **mammals** and **food** with **VGG16** (ImageNet-pretrained, torchvision). The embedding is the penultimate fully-connected layer **`fc2` (4096-dim)**

In this script we extract features for all bird images using the pretrained `VGG16` net contained in the `keras` module and by deleting the output layer. 

```python
import numpy as np
import pandas as pd
import keras
import os
import scipy

from keras.models import Model
from tensorflow.keras import applications
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input,decode_predictions

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
from scipy.stats import pearsonr
```

```python
# load the pre-trained model
vgg16 = keras.applications.VGG16(weights='imagenet', include_top=True, pooling='max', input_shape=(224, 224, 3))
# set the fc2 layer as output
basemodel = Model(inputs=vgg16.input, outputs=vgg16.get_layer('fc2').output)
```

```python
def img_preprocessing(img):
  img_array = image.img_to_array(img)
  img_array = np.expand_dims(img_array, axis=0)
  img_array = preprocess_input(img_array)
  return img_array
```

```python
def get_features(img_array):
  vector = basemodel.predict(img_array)
  return vector
```

## Mammals

```python
n_stim     = 80
n_features = 4096 # based on VVG16


# get a list of all files in the 'Transparent' folder
files = os.listdir('../../Materials/images/Mammals/')[:-1]

feature_matrix = np.empty(shape=(n_stim, n_features))

# loop through all the image files in the files list
for img_file in files:

  # load image
  img = image.load_img(os.path.join('../../Materials/images/Mammals/', img_file), color_mode='rgb', target_size=(224, 224))

  # preprocess image
  img_array = img_preprocessing(img)

  # get feature vector 
  features = get_features(img_array)

  # get current index
  i = files.index(img_file)

  # add vector to matrix at current index
  feature_matrix[i,] = features

```

```python
# Normalize the feature matrix
feature_matrix = feature_matrix / np.linalg.norm(feature_matrix, axis=1, keepdims=True) 
```

```python
# save the dataframe as a csv file
pd.DataFrame(feature_matrix).to_csv("../../Data/Embeddings/image_embeddings_mammals_vgg16.csv")
```
